Feature Selection Part

```python
animes.columns
```

```python
animes=animes[['id','title','genre','overview','image']]
```

```python
animes
```

```python
animes['tags']=animes['overview']+animes['genre']
```

```python
animes
```

```python
new_data=animes.drop(columns=['overview','genre'])
```

```python
new_data
```

```python
new_data.columns
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv=TfidfVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Trigun"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(animes):
    index=new_data[new_data['title']==animes].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Trigun")
```
